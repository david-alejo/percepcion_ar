# Práctica de Percepción: Procesamiento de Imágenes y Geometría
## Introducción
En esta práctica veremos operaciones básicas a realizar sobre imágenes. Para ello se utilizará OpenCV, NumPy y Matplotlib.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def imshow(img, title=""):
    if len(img.shape)==2:
        plt.imshow(img, cmap="gray")
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
```

## Ejercicio 1: Operaciones a nivel de píxel y variación de histograma
Objetivos: Manipular intensidades de píxel, modificar contraste y brillo, y analizar histograma en escala de grises.


```python
# Carga de imagen
img = cv2.imread("./images/Bici.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Variación de brillo y contraste
alpha = 1.3  # contraste
beta = 25    # brillo
mod = cv2.convertScaleAbs(gray, alpha=alpha, beta=beta)

# --- VISUALIZACIÓN COMPARATIVA ---
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
imshow(gray, "Original")

plt.subplot(1,2,2)
imshow(mod, "Brillo+Contraste")

plt.tight_layout()
plt.show()

# --- HISTOGRAMAS ---
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title("Histograma original")

plt.subplot(1,2,2)
plt.hist(mod.ravel(), bins=256, range=[0,256])
plt.title("Histograma modificado")

plt.tight_layout()
plt.show()
```

## Ejercicio 2: Histograma en RGB
Objetivos: analizar el histograma en cada canal, aplicando filtrado en cada canal.


```python
img = cv2.imread("./images/Lego.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

colors = ("b", "g", "r")

fig, axes = plt.subplots(1, 2, figsize=(12,5))

# --- IZQUIERDA: imagen ---
axes[0].imshow(img_rgb)
axes[0].set_title("Imagen")
axes[0].axis("off")

# --- DERECHA: histogramas ---
for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0,256])
    axes[1].plot(hist, color=col)

axes[1].set_title("Histogramas RGB")
axes[1].set_xlim([0,256])

plt.tight_layout()
plt.show()
```

## Ejercicio 3: filtrado de imágenes
Objetivo: comparar diferentes filtros.


```python
# Cargar imagen
img = cv2.imread("./images/balloons_noisy.png")

# Aplicar filtros
box = cv2.blur(img,(7,7))
gauss = cv2.GaussianBlur(img,(7,7),0)
median = cv2.medianBlur(img,7)

# Mostrar resultados
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
imshow(img,"Original")

plt.subplot(2,2,2)
imshow(box,"Box")

plt.subplot(2,2,3)
imshow(gauss,"Gaussian")

plt.subplot(2,2,4)
imshow(median,"Mediana")


plt.tight_layout()
plt.show()
```
